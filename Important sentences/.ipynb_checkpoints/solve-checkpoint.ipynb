```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import pymorphy2 as ppy
import re
import sklearn.feature_extraction.text as sktext
```

```python
x = pd.read_csv('data/test_1.csv', encoding="UTF-8")
```

```python
x.religion/=10
x.world/=10
x.science/=10
x.culture/=10
x.economy/=10
x.politics/=10
x.society/=10
```

```python
train = x[(x.religion>0) | (x.world>0) | (x.science>0) | (x.culture>0) | (x.economy>0) | (x.politics>0) | (x.society>0)]
```

```python
morph = ppy.MorphAnalyzer()
```

```python
vectorizer = sktext.CountVectorizer()
title_vectorizer = sktext.CountVectorizer()
X_texts_tests = vectorizer.fit_transform(x.text)
X_titles_tests = title_vectorizer.fit_transform(x.title)
X_texts_train = vectorizer.transform(train.text)
X_titles_train = title_vectorizer.transform(train.title)
```

```python
train_y = train.drop(['Unnamed: 0','title', 'text'], axis=1)
```

```python
train_y.head()
```

```python
from catboost import CatBoostRegressor
from sklearn.cross_validation import ShuffleSplit, cross_val_score
from sklearn.linear_model import SGDRegressor
from sklearn.grid_search import GridSearchCV
```

# Линейный регрессор

```python
clf = CatBoostRegressor()

my_ans = []
clf.fit(X_texts_train.toarray(), train_y.world)
mat1 = clf.predict(X_texts_tests.toarray()).tolist()
my_ans = mat1
for i in range(len(my_ans)):
    f = my_ans[i] 
    my_ans[i] = list()
    my_ans[i].append(f)
clf.fit(X_texts_train.toarray(), train_y.science)
mat1 = clf.predict(X_texts_tests.toarray()).tolist()

clf.fit(X_texts_train.toarray(), train_y.culture)
mat2 = clf.predict(X_texts_tests.toarray()).tolist()

clf.fit(X_texts_train.toarray(), train_y.economy)
mat3 = clf.predict(X_texts_tests.toarray()).tolist()

clf.fit(X_texts_train.toarray(), train_y.politics)
mat4 = clf.predict(X_texts_tests.toarray()).tolist()

clf.fit(X_texts_train.toarray(), train_y.society)
mat5 = clf.predict(X_texts_tests.toarray()).tolist()

clf.fit(X_texts_train.toarray(), train_y.religion)
mat6 = clf.predict(X_texts_tests.toarray()).tolist()

for i in range(len(my_ans)):
    my_ans[i].append(mat1[i])
    my_ans[i].append(mat2[i])
    my_ans[i].append(mat3[i])
    my_ans[i].append(mat4[i])
    my_ans[i].append(mat5[i])
    my_ans[i].append(mat6[i])
```

```python
maxs = []
maxsi = []
for i in range(len(my_ans)):
    max1 = max(my_ans[i][0], my_ans[i][1])
    max2 = min(my_ans[i][0], my_ans[i][1])
    if (max1==my_ans[i][0]):
        max1i = 0
        max2i = 1
    else:
        max1i = 1
        max2i = 0
    for j in range(2,7):
        if my_ans[i][j]>max1:
            max2 = max1
            max2i = max1i
            max1i = j
            max1 = my_ans[i][j]
        elif my_ans[i][j]>max2:
            max2 = my_ans[i][j]
            max2i = j
    maxs.append(max1)
    maxsi.append(max1i+1)
    print(i, ' - [',max1i+1,'-',max1,' ',max2i+1,'-',max2,'],')
```

```python
CAEGORIES = ['world', 'science', 'culture', 'economy', 'politics','society','religion']
with open('y.csv', 'w', newline='') as y:
    first_col = 'category'
    writer = csv.writer(y, delimiter=' ')
    writer.writerow([first_col])
    for item in maxsi:
        writer.writerow([CAEGORIES[item-1]])
```
